# Handling missing values


```python
churn_data["Churn Category"].fillna("Not Mentioned", inplace= True)
churn_data["Churn Reason"].fillna("Not Mentioned", inplace= True)
```

```python
churn_data
```

```python
churn_data.isnull().sum()
```

```python
cols = churn_data.columns
null_cols = []
for column in cols:
  if churn_data[column].isnull().values.any() == True:
    #print(column, churn_data[column].isnull().values.sum())
    null_cols.append(column)
print(null_cols)
```

```python
for column in null_cols:
  perm = np.random.permutation(churn_data[column])
  nan_val = np.argwhere(perm != perm)
  perm = np.delete(perm, nan_val)
  end = len(perm)
  churn_data[column] = churn_data[column].apply(lambda z: perm[np.random.randint(end)] if pd.isnull(z) else z)
```

```python
churn_data.isnull().sum()
```

```python
cols_to_encode=churn_data.dtypes[churn_data.dtypes == "object"].index.values.tolist()

cols_to_encode
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 

for col in cols_to_encode:
    churn_data[col]= label_encoder.fit_transform(churn_data[col]) 

    print(churn_data[col].unique())
```

```python
cloumn_drop = ['Customer Status','Age Group']

X = churn_data.drop(columns=cloumn_drop,axis=1)
Y = churn_data['Customer Status']

from sklearn.model_selection import train_test_split

# Xtrain,Xtest,Ytrain,y_test = train_test_split(X,Y, train_size=0.7, random_state=42)
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,Y, test_size=0.2,random_state=42)
Xtrain.shape,Ytrain.shape,Xtest.shape,Ytest.shape
```

```python
DT_classifier = DecisionTreeClassifier()

DT_classifier.fit(Xtrain, Ytrain)
dt_pred = DT_classifier.predict(Xtest)
print("RandomForestClassifier:")
print("Accuracy:", accuracy_score(Ytest, dt_pred))
print("Precision:", precision_score(Ytest, dt_pred,average=None))
print("Recall:", recall_score(Ytest, dt_pred,average=None))
print("F1-score:", f1_score(Ytest, dt_pred,average=None))

print("Confusion Matrix: ", 
        confusion_matrix(Ytest, dt_pred))
```

# Resampling method

```python
from sklearn.utils import resample
data_maj = churn_data[(churn_data['Customer Status']==2)] 
data_min = churn_data[(churn_data['Customer Status']==0)] 
data_min2 = churn_data[(churn_data['Customer Status']==1)] 

data_min_upsampled = resample(data_min,replace=True,n_samples= 4720,random_state=42)
data_min_upsampled2 = resample(data_min2,replace=True,n_samples= 4720,random_state=42)

balanced_churn_data = pd.concat([data_min_upsampled,data_min_upsampled2, data_maj])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical


column_drop = ['Customer Status', 'Age Group']
X = balanced_churn_data.drop(columns=column_drop, axis=1)
Y = balanced_churn_data['Customer Status']


Y_encoded = to_categorical(Y, num_classes=3)

X_train, X_test, y_train, y_test = train_test_split(X, Y_encoded, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = Sequential()
model.add(Dense(128, activation='relu', input_dim=X_train.shape[1]))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax')) 
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history=model.fit(X_train_scaled, y_train, epochs=30, batch_size=32, validation_split=0.2)

y_pred_probs = model.predict(X_test_scaled)
y_pred = [int(round(prob.argmax())) for prob in y_pred_probs]

print(f'Accuracy: {accuracy_score(y_test.argmax(axis=1), y_pred)}')
print(classification_report(y_test.argmax(axis=1), y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score

def plot_confusion_matrix(ax, y_true, y_pred, labels, title):
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels, ax=ax)
    ax.set_title(title)
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

def plot_classification_report(ax, y_true, y_pred, target_names, title):
    report = classification_report(y_true, y_pred, target_names=target_names, output_dict=True)
    df_report = pd.DataFrame(report).transpose()
    sns.heatmap(df_report.iloc[:-1, :].T, annot=True, cmap='Blues', fmt=".3f", ax=ax)
    ax.set_title(title)

y_pred_rf = RF_classifier.predict(Xtest)
y_pred_dt = DT_classifier.predict(Xtest)
y_pred_gb = clf.predict(Xtest)
y_true_ml = Ytest

target_names_ml = ['Class 0', 'Class 1', 'Class 2']

fig, axs = plt.subplots(3, 2, figsize=(10, 16))

plot_confusion_matrix(axs[0, 0], y_true_ml, y_pred_rf, target_names_ml, 'Random Forest Result (Confusion Matrix)')
plot_classification_report(axs[0, 1], y_true_ml, y_pred_rf, target_names_ml, 'Random Forest Result (Classification Report)')

plot_confusion_matrix(axs[1, 0], y_true_ml, y_pred_dt, target_names_ml, 'Decision Tree Result (Confusion Matrix)')
plot_classification_report(axs[1, 1], y_true_ml, y_pred_dt, target_names_ml, 'Decision Tree Result (Classification Report)')

plot_confusion_matrix(axs[2, 0], y_true_ml, y_pred_gb, target_names_ml, 'Gradient Boosting Result (Confusion Matrix)')
plot_classification_report(axs[2, 1], y_true_ml, y_pred_gb, target_names_ml, 'Gradient Boosting Result (Classification Report)')

plt.tight_layout()
plt.show()
```

```python
def plot_cross_validation_results(results, model_names):
    plt.figure(figsize=(10, 6))
    plt.bar(model_names, results.mean(axis=1), yerr=results.std(axis=1), color=['skyblue', 'lightgreen', 'lightcoral'])
    plt.title('Cross-Validation Results for Machine Learning Models')
    plt.ylabel('Accuracy')
    plt.show()


results_rf = cross_val_score(RF_classifier, X, Y, cv=5, scoring='accuracy')
results_dt = cross_val_score(DT_classifier, X, Y, cv=5, scoring='accuracy')
results_gb = cross_val_score(clf, X, Y, cv=5, scoring='accuracy')

plot_cross_validation_results(np.array([results_rf, results_dt, results_gb]), ['Random Forest', 'Decision Tree', 'Gradient Boosting'])
```
